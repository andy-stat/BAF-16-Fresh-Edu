# 4.결측치 처리

```python
train.isna().sum() # 결측치 비율이 매우 작으므로 결측치 버리기
train = train.dropna()
```

# 5.범주형 변수 인코딩

```python
# 특성에 순서가 존재하므로 Basic 0 <<  Standard	1 <<  Premium 2 로 라벨 인코딩
train['Subscription Type'].value_counts()
train['Subscription Type'] = train['Subscription Type'].map({'Basic':0,'Standard':1,'Premium':2})
```

```python
# Gender의 경우 원핫 인코딩
train['Gender']=train['Gender'].map({"Female":0,'Male':1})
```

```python
# 이 경우에도 원핫인코딩
print(train['Contract Length'].value_counts()) # 여기에서 Monthly데이터는 제거
train_no = train[~(train['Contract Length']=='Monthly')]
train_no['Contract Length'] = train_no['Contract Length'].map({"Quarterly":0,'Annual':1})
train_no['Contract Length'].value_counts() # 바뀐것 확인
```
